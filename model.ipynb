# Modelling and Testing
- Over here, the training data will go through preprocessing before entering a model to be trained
- From there, we will let the test set go in and see what the values and check its score 

```python
# Import training set
import pandas as pd

data = pd.read_csv('data/train.csv')
```

```python
data
```

```python
#Split features and target
X = data.drop(columns='NObeyesdad')
y = data['NObeyesdad']
```

```python
X
```

```python
#Scaling and Encoding

#Numerical
from sklearn.preprocessing import MinMaxScaler

#Categorical
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

num_cols = X.drop(columns='id').select_dtypes(include = ['number']).columns
cat_cols_ord = ['family_history_with_overweight', 'FAVC', 'SCC', 'CALC']
cat_cols_ohe = ['CAEC', 'SMOKE', 'MTRANS']
```

```python
#ColumnTransformer
from sklearn.compose import ColumnTransformer
```

```python
#Column transforming
preproc = ColumnTransformer(
    transformers = [
        ('minmax', MinMaxScaler(), num_cols),
        ('ordinal', OrdinalEncoder(), cat_cols_ord),
        ('onehot', OneHotEncoder(handle_unknown='ignore'), cat_cols_ohe),
    ], remainder = 'drop'
)
```

```python
#initiate model and pipeline
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
from catboost import CatBoostClassifier
import xgboost as xgb

pipeline1 = Pipeline(steps =[
    ('preprocess', preproc),
    ('model', GradientBoostingClassifier())
])
```

```python
pipeline2 = Pipeline(steps =[
    ('preprocess', preproc),
    ('model', CatBoostClassifier())
])
```

# Based on the results of the two models
- Grad Boost is the better model 

# Train on entire data
- Train on the entire set
- Predict the test set
- Submit CSV file

```python
X_transformed = preproc.fit_transform(X)
```

```python
model = GradientBoostingClassifier()
model.fit(X_transformed,y)
```

```python
test_set = pd.read_csv('data/test.csv')
```

```python
test_set
```

```python
test_answer =  model.predict(preproc.fit_transform(test_set))
```
